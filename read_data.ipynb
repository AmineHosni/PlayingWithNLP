```python
file_base_name = "data/House Debates Testing Sets/hansard.36.2.house.debates."
```

```python
import pandas as pd
```

```python
def read_data(lang = 'e'):
    ####################
    # e: english
    # f: french
    ####################
    series = pd.Series()
    for i in range(58,95):
        lines = open("{}{:03d}.{}".format(file_base_name,i,lang)).read().splitlines()
        series = series.append(pd.Series(lines))
    return series
```

```python
serie_english = read_data()
```

```python
serie_french = read_data(lang='f')
```

```python
import spacy

nlp = spacy.load('fr_core_news_sm')

# for sent in serie_french:
#     print(sent)
df_fr = pd.DataFrame(data=serie_french,columns=['text'])
df_fr.head()
```

```python
from pprint import pprint
```

```python
def remove_PER(sentence):
    s = nlp(sentence)
    new_sentence = ' '.join(token.text for token in s.ents if token.label_ not in ['PER'])
    return new_sentence
```

```python
remove_PER(df_fr.iloc[0]['text'])
```

```python
df_fr['text_/PER'] = df_fr['text'].apply(remove_PER)
```

```python
df_fr.head()
```

```python
s = nlp("Amine Hosni est un data scientiste passioné et impliqué")
' '.join(token.text for token in s if token.pos_ not in ['PROPN'] )
```

```python
def remove_PROPN(sentence):
    s = nlp(sentence)
    new_sentence = ' '.join(token.text for token in s if token.pos_ not in ['PROPN'] )
    return new_sentence

df_fr['text_/PROPN'] = df_fr['text'].apply(remove_PROPN)
```
